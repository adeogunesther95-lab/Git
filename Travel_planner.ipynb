```python
from rich import markdown
from rich import print
from rich.markdown import Markdown
import requests

def display_weather(location):
  """Display weather forecast for origin and destination"""
  api_key = "bf602aabco34t729377499af62121a7a"
  api_url = f"https://api.shecodes.io/weather/v1/current?query={location}&key={api_key}"

  response = requests.get(api_url)
  data = response.json()



  temperature = round(data['temperature']['current'])
  description = data['condition']['description']

  print(f"The current weather in [bold]{location}[/bold] is {description} with a temperature of [bold]{temperature}°C[/bold]")

def generate_itinerary(origin, destination, duration):
  """Generate travel itinery from origin to destination for the next duration days"""
  prompt = f"Generate a travel itinerary from {origin} to {destination} for the next {duration} days.Please state the different transportation means available for the trip together with any side attractions, and recommend the best transportation means. Please include an estimated cost of the trip for each day."
  context = "You are an experienced travel agent. Please try to be as concise as possible"
  api_key = "bf602aabco34t729377499af62121a7a"
  api_url = f"https://api.shecodes.io/ai/v1/generate?prompt={prompt}&context={context}&key={api_key}&units=metric"

  response = requests.get(api_url)
  data = response.json()
  markdown_text = Markdown(data['answer'])
  print(markdown_text)

origin = input("Enter the city of originination: ").strip().title()
destination = input("Enter the city of destination: ").strip().title()
duration = input("How many days will your trip last? (enter a number only, i.e. 1,2 etc.) ").strip()

print(f"\n\nGenerating travel itinerary from {origin} to {destination} for the next {duration} days...\n")

def welcome():
  print("[bold magenta]Welcome to the Travel Itinerary Planner![/bold magenta]")

def credits():
  print(f"\n\n[magenta]Thank you for using my travel itinerary planner ❤, built by [bold]Esther Adeogun[/bold][/magenta]\n")

welcome()

if origin and destination and duration.isdigit():
  display_weather(origin)
  display_weather(destination)
  generate_itinerary(origin, destination, duration)
  credits()
else:
  print("Something is wrong, please complete all fields and try again")
```
